### Stock Analysis Usage Examples


```python
import utils.analyser as an
import pandas as pd

# Pull data using DataUtils
dl = DataUtils()
s1_name = "NVDA"
s2_name = "AAPL"
start_date = "2014-01-01"
end_date = "2021-01-01"

s1 = dl.get_data(s1_name, start_date, end_date)
s2 = dl.get_data(s2_name, start_date, end_date)
```

## Stationarity & Cointegration Tests


```python
def printc(text):
    print(f"\n\n\n\n{'-'*15} {text} {'-'*15}")


# Check stationarity
printc("Check Stationarity with ADF Test")
p_value = an.check_stationarity(s1, verbose=True)
print(f"P-value: {p_value}")

# Plot Stationary Analysis
printc("Plot Stationary Analysis")
an.plot_stationary_analysis(s1)

# Transform to stationary
printc("Transform to Stationary")
stationary_data = an.transform_to_stationary(s1, verbose=True)

display(stationary_data.tail(5))
an.subplot_lines(s1, stationary_data, "Stationary Transformation")

# Plot seasonal decomposition
printc("Plot Seasonal Decomposition")
an.plot_seasonal_decomposition(s1)

# Test Granger causality
printc("Test Granger Causality")
causality = an.test_granger_causality(s2, s1)
display(causality)

# Calculate cointegration matrix
printc("Calculate Cointegration Matrix")
cointegration_matrix = an.calculate_cointegration_matrix(pd.concat([s1, s2], axis=1))
display(cointegration_matrix)

# Identify cointegrated pairs
printc("Identify Cointegrated Pairs")
score_matrix, p_value_matrix, significant_pairs = an.identify_cointegrated_pairs(
    pd.concat([s1, s2], axis=1)
)
print(f"Score Matrix")
display(score_matrix)
print(f"Ranked Significant pairs")
display(significant_pairs)
```

### Rolling window analysis


```python
# Calculate moving average
printc("Calculate Moving Average")
WINDOW = 60
moving_avg = an.calculate_moving_average(s1, window=WINDOW)
display(moving_avg.tail(5))
an.subplot_lines(s1, moving_avg, "Moving Average")

# Calculate exponential moving average
printc("Calculate Exponential Moving Average")
WINDOW = 60
ema = an.calculate_exponential_moving_average(s1, window=WINDOW)
display(ema.tail(5))
an.subplot_lines(s1, ema, "Exponential Moving Average")

# Apply rolling OLS
printc("Apply Rolling OLS")
WINDOW = 60
rolling_ols = an.apply_rolling_ols(s1, s2, window=WINDOW)
display(rolling_ols.tail(5))
an.subplot_lines(s1, rolling_ols, "Rolling OLS")


# Calculate rolling log return z-score
print("\nCalculate Rolling Log Return Z-Score")
WINDOW = 60
spread, zscore = an.rolling_logret_zscore_nb(s1, period=WINDOW)
spread = spread[1:]
zscore = zscore[WINDOW:]
spread = (spread - spread.mean()) / spread.std()
zscore = (zscore - zscore.mean()) / zscore.std()

display(spread)
display(zscore)
plt.plot(zscore, label="Z-Score")
plt.plot(spread, label="Spread")
plt.title("Rolling Log Return Z-Score")
plt.legend()
plt.show()


# Calculate rolling OLS z-score
print("\nCalculate Rolling OLS Z-Score")
spread, zscore = an.rolling_ols_zscore_nb(s1, period=20)
```

### Usage examples of additional functions related to stock time series analysis


```python
# Calculate Z-score
printc("Calculate Z-score")
zscore = an.calculate_zscore(s1)
display(zscore.tail(5))

# Calculate volatility
printc("Calculate Volatility")
volatility = an.calculate_volatility(s1)
display(volatility.tail(5))
an.subplot_lines(s1, volatility, "Volatility")

# Calculate OLS spread
print("\nCalculate OLS Spread")
spread = an.ols_spread_nb(s1)
print(f"Spread: {spread}")

# Calculate Bollinger Bands
printc("Calculate Bollinger Bands")
bollinger_bands = an.calculate_bollinger_bands(s1)
display(bollinger_bands.tail(5))
an.subplot_lines(s1, bollinger_bands, "Bollinger Bands")

# Calculate Relative Strength Index (RSI)
printc("Calculate Relative Strength Index (RSI)")
rsi = an.calculate_relative_strength_index(s1)
display(rsi.tail(5))
an.subplot_lines(s1, rsi, "RSI")

# Calculate Pearson correlation
printc("Calculate Pearson Correlation")
correlation = an.calculate_pearson_correlation(s1, s2)
print(f"Pearson Correlation: {correlation:.4f}")
```
